We have IQR = Q3−Q1

Lower Bound=Q1−1.5×IQR

Upper Bound=Q3+1.5×IQR


```python
Q1 = df['Annual Income'].quantile(0.25)
Q3 = df['Annual Income'].quantile(0.75)
IQR = Q3 - Q1

# Calculate the lower and upper bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

```python
df_filtered = df[(df['Annual Income'] >= lower_bound) & (df['Annual Income'] <= upper_bound)]

print(len(df))                    # Original number of rows
print(len(df_filtered))           # Number of rows after removing outliers
```

```python
plt.figure(figsize=(8, 6))
plt.boxplot(df_filtered['Annual Income'])
plt.title('Box Plot of Annual Income')
plt.ylabel('Annual Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.hist(df_filtered['Annual Income'], bins=50, edgecolor='black')
plt.title('Histogram of Annual Income')
plt.xlabel('Annual Income')
plt.ylabel('Frequency')
plt.show()
```

```python

sns.boxplot(x='Body Style', y='Annual Income', data=df_filtered)
plt.title('Annual Income vs. Body Style')
plt.show()
```
